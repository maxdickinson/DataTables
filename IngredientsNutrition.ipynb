```python

def percToNumMagn(entry):
    dailyValueMagn= 420

    #takes a string that has a % daily value and uses the total
    #daily value to get the amount in the ingredient
    #only applies to the column that is the ingredient
    #for the daily value
    if type(entry) == str:
        entry = str.lower(entry)
        if '%' in entry:
            numPercent = entry.split('%')
            num1 = int(numPercent[0])
            amount = (num1/100)*dailyValueMagn

            return amount
    else: return entry

ingredNutrit.iloc[:,23] = ingredNutrit.iloc[:,23].apply(percToNumMagn)
```
