```python
# goes from a csv to a frame 
def newRecipeGetNutrtionAddToFrames(ingredientXNutrition2 , recipeXIngredients , recipeXNutrition2, ):
    #load ingredient nutrition clean csv 
    #intedended to be run in dataFrame folder from juptyyre notebook 
    #rename index as ingredients
    ingredientXNutrition2 = ingredientXNutrition2.rename({'Unnamed: 0' : 'ingredients'} , axis =1)
    ingredientXNutrition2 = ingredientXNutrition2.set_index('ingredients')
    
    recipeXIngredients =  recipeXIngredients.set_index('RecipeDateType')

    recipeXNutrition2.set_index('RecipeDateType')

    # create a new frame with the ingredients as the columns to enter recipes in
    ingredients = ingredientXNutrition2.index.to_list()
    #recipe frame is the one where a new recipe and its ingredients will go
    recipeFrame = pd.DataFrame(np.zeros((1,len(ingredients))))
    recipeFrame = recipeFrame.set_axis(ingredients,axis = 1)
    

    #data types for new recipe frame 
    teaspoonIngreds = ['bakingSoda(teaspoons)',
    'bakingPowder(teaspoons)',
    'cinamon(teaspoons)',
    'nutmeg(teaspoons)',
    'cardamom(teaspoons)',
    'xanthamGum(teaspoons)',
    'Salt(teaspoons)',
    'guarGuam(teaspoons)',
    'cornStarch(teaspoons)',
    'vanillaExtract(teaspoons)',
    'Almond Extract(teaspoons)',
    'naturesFlavorsButterExtract(teaspoons)',
    'espressoPowder(teaspoons)',]
    teaFloat = dict(zip(teaspoonIngreds,13*[float]))
    #change datatypes of recipe frame to int and float where teaspoons or not
    recipeFrame = recipeFrame.astype(float)
    recipeFrame = recipeFrame.astype(teaFloat)
    #get the input from user of the different amounts of ingredients used in the recipe 
    #need to enter whole numbers, decimals or fractions, gives an error for any strings
    #escapes and enters entered as nans 
    

    #getting input of ingredients for new recipe
    for i in np.arange(len(ingredients)):
        if ingredients[i] in teaspoonIngreds:
            inhere = input('How many teaspoons of ' + ingredients[i])
            try:
                recipeFrame.iloc[0,i] = pd.to_numeric(inhere)
                
            except:
                recipeFrame.iloc[0,i] = pd.to_numeric(pd.eval(inhere))
                
            
        else:
            inhere2 = input('How many grams of ' + ingredients[i])
            try:
                recipeFrame.iloc[0,i] = pd.to_numeric(inhere2)
                
                
            except:
                recipeFrame.iloc[0,i] = pd.to_numeric(pd.eval(inhere2))
    #better name for frame
    recipeIngredientFrame = recipeFrame.fillna(0)

    #take prodcut between the recipe series and the ingredient with nutrition values frame
    
    recipeDateType = input('What is the recipe date and type of cookie')


    #make nutritional values with matrix multiplication
    nutritionalValuesFrame = recipeIngredientFrame.dot(ingredientXNutrition2)
    #nutritionalValuesFrame = nutritionalValuesFrame.set_axis([recipeDateType] , axis = 0)

    #add recipe to top of frame
    #recipeNutIndex = [recipeDateType] + recipeXNutrition2.iloc[:,-1].to_list() 
    recipeXNutrition2 = pd.concat([nutritionalValuesFrame , recipeXNutrition2] , axis =0)
    recipeXNutrition2.iloc[0,-1] = recipeDateType
    recipeXNutrition2 = recipeXNutrition2.set_index('RecipeDateType')

    print(nutritionalValuesFrame)
    print(nutritionalValuesFrame/5)
    #fiver = nutritionalValuesFrame.div(5)
    #print('Values with 5 Cookies' , fiver)

    #grab date and type of recipe and put it in the index 
    recipeIngredientFrame = recipeIngredientFrame.set_axis([recipeDateType] , axis = 0)
    #insert new row at top of big frame with all recipes and ingredients

    recipeXIngredients = pd.concat([recipeIngredientFrame,recipeXIngredients] , axis = 0)
    #print(recipeXNutrition2.info())
    #print(recipeXNutrition2.info())
    # recipeXNutrition2.insert(1,'Amount per 1 cookie, if 5 in batch' ,recipeXNutrition2.iloc[:,0].div(5))
    
    return ingredientXNutrition2 , recipeXIngredients , recipeXNutrition2
        
```

```python
ingredientXNutritionFrame , recipeXIngredientsFrame , recipeXNutritionFrame = newRecipeGetNutrtionAddToFrames(ingredientXNutrition3 ,recipeXIngredients2 ,recipeXNutrition3)
```
